# Sampling rate benchmark 

```python
import numpy as np
import matplotlib.pyplot as plt

song_duration = 30
sampling_rates = np.linspace(start=25, stop=50000, num=500)
eq_sq_img_size = np.round(np.sqrt((song_duration*sampling_rates/3)))

left_lim = 0
right_lim = sampling_rates[-1]+1
bottom_lim = 0
top_lim = eq_sq_img_size[-1]+1
 

plt.figure(figsize=(16,8))
plt.plot(sampling_rates, eq_sq_img_size, c='midnightblue')
plt.vlines(x=44100, ymin=bottom_lim, ymax=top_lim, label="CD standard", linestyle='--', color='r')
plt.hlines(y=112, xmin=left_lim, xmax=right_lim, label="ArcFace training", linestyle='--', color='darkred')

plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor', linestyle='--')

plt.suptitle("To what size of square image does a 30s song extract correspond to?")
plt.title("(Given same bit depth)")

plt.xlabel("Song sampling rate")
plt.xlim(left_lim, right_lim)

plt.ylabel("Image length")
plt.ylim(bottom_lim, top_lim)

plt.legend(loc="lower right")
plt.show()
```

```python
for fixed_image_size in [112, 128, 256, 512, 664]:
    corresponding_sampling_freq = np.round(sampling_rates[np.argmin(np.abs(eq_sq_img_size-fixed_image_size))])
    print(f"Nearest sampling frequency for a {fixed_image_size}x{fixed_image_size} image: {corresponding_sampling_freq} Hz")
```
